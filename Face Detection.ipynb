```python
import cv2
import PIL

face_cascade = cv2.CascadeClassifier('readonly/haarcascade_frontalface_default.xml')

img = cv2.imread('readonly/0a-0.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = (face_cascade.detectMultiScale(gray, 1.235, 6))
ct = contact_sheet(len(faces))
```

```python
from PIL import Image, ImageDraw

image = Image.open('readonly/0a-0.png').convert('RGB')
d = ImageDraw.Draw(image)
x = 0
for face in faces:
    box = [face[0],face[1],face[0]+face[2],face[1]+face[3]]
    img = image.crop(box).resize((125, 125), PIL.Image.LANCZOS)
    ct.paste(img,(x,0))
    x += 125

display(ct)
```

```python
import PIL
from PIL import Image

def contact_sheet(ct):
    c_sheet = PIL.Image.new('RGB', (125*ct, 125))
    return c_sheet
```
